```python
import numpy as np
```

```python
def winner(rank_a, rank_b):
    total = rank_a + rank_b
    if np.random.uniform() > rank_a/total:
        winner = rank_a    
    else:
        winner = rank_b
        
    print('{} seed wins!'.format(int(winner)))
    return winner
```

```python
def full_bracket(round1):
    np.random.seed(2023)
    f, s, t, q, m = 0, 0, 0, 0, 0
    round2, round3, round4, semi, final = [], [], [], [], []
    print('First round:')
    while f<64:
        round2_add = winner(round1[f], round1[f+1])
        round2 = np.append(round2, round2_add)
        f += 2
    
    print('\nSecond Round:')
    while s<32:
        round3_add = winner(round2[s], round2[s+1])
        round3 = np.append(round3, round3_add)
        s += 2
        
    print('\nThird Round:')
    while t<16:
        round4_add = winner(round3[t], round3[t+1])
        round4 = np.append(round4, round4_add)
        t += 2
        
    print('\nQuarterfinals:')
    while q<8:
        semi_add = winner(round4[q], round4[q+1])
        semi = np.append(semi, semi_add)
        q += 2
        
    print('\nSemifinals:')
    while m<4:
        final_add = winner(semi[m], semi[m+1])
        final = np.append(final, final_add)
        m += 2
    
    print('\nFinal:')
    champion = winner(final[0], final[1])
```

```python
round1 = [1, 16, 8, 9, 5, 12, 4, 13, 6, 11, 3, 14, 7, 10, 2, 15] * 4

full_bracket(round1)
```
